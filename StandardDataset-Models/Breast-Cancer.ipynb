```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

cancer = load_breast_cancer()
type(cancer)
```

```python
print("Keys for cancer:\n{}".format(cancer.keys()))
print("Shape of data:\n{}".format(cancer["data"].shape))
```

```python
print(
    "Sample count per class:\n{}".format(
        {n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}
    )
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, random_state=66, stratify=cancer.target
)
print("Train and test split data shape: \n{},{}".format(X_train.shape, X_test.shape))
```

```python
training_accuracy = []
test_accuracy = []
neighbors_setting = range(1, 11)

for k in neighbors_setting:
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_setting, training_accuracy, label="training accuracy")
plt.plot(neighbors_setting, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression

linreg = LinearRegression().fit(X_train, y_train)
logreg = LogisticRegression().fit(X_train, y_train)
logreg_c = LogisticRegression(C=0.0100).fit(X_train, y_train)
print("Training set score: {:.3f}".format(linreg.score(X_train, y_train)))
print("Test set score: {:.3f}".format(linreg.score(X_test, y_test)))
print("Training set score: {:.3f}".format(logreg.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg.score(X_test, y_test)))
print("Training set score: {:.3f}".format(logreg_c.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg_c.score(X_test, y_test)))
```
